# Data Cleaning

## 1. Handling MIssing Values

```python
# check how many null values are there in the data 
data.isnull().sum()
```

```python
# to decide whether the feature should be dropped or not 
print((data.isnull().sum() / len(data)) * 100)
```

#### for features with missing values > 50% i.e cabin, let's investigate 

```python
# 1. understand cabin data 
data["Cabin"].head(10)
```

```python
data["Cabin"].mode()
```

```python
# making a copy of the data to avoid changes in the original data 
df = data
```

```python
# if we rethink why we need cabin, we realise that we only need deck (lower deck, upper deck) cabin doesn't matter so extract deck from cabin
df["Deck"] = df["Cabin"].str[0]
```

```python
df.head()
```

```python
# now fill the missing values of the deck with "unknown"
df["Deck"] = df["Deck"].fillna("Unknown")
```

```python
df['Deck'].unique()
```

```python
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
df["Embarked"].isnull().sum()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
df.drop("Cabin", axis= 1, inplace= True)
```

```python
df.isnull().sum()
```

## 4. Multivariate Analysis

1. Sex and Age

```python
sns.boxplot(data=df, x= "Sex", y= "Age", hue= "Survived")
```

```python
df["AgeGroup"] = pd.cut(
    df["Age"],
    bins = [0, 12, 18, 35, 60, 100],
    labels = [
        'Child',
        'Teen',
        'YoungAdult',
        'Adult',
        'Senior'
    ]
)
```

```python
sns.catplot(data= df, x = "AgeGroup", col = "Sex", hue= "Survived", kind= 'count')
```

```python
pd.crosstab(
    [df["Sex"], df["AgeGroup"]],
    df["Survived"],
    normalize= 'index'
)*100
```

2. Family Size and Deck

```python
def family_count(size):
    if size == 1:
        return "alone"
    elif size <= 4:
        return "Small Family"
    else:
        return "Large Family"


df["familysize"] = df["familysize"].apply(family_count)
```

```python
sns.catplot(data=df, x= "Deck", col= "familysize", hue= "Survived", kind= "count")
```

3. Embarked and Pclass

```python
sns.catplot(data= df, x = "Embarked", col = "Pclass", hue= "Survived", kind= "count")
```

4. Sex and Pclass

```python
sns.catplot(data= df, x = "Sex", col = "Pclass", hue= "Survived", kind= 'count')
```

5. Deck and Fare

```python
sns.boxplot(data= df, x = "Deck", y = "Fare", hue= "Survived")
```
